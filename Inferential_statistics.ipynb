# Inferential Statistics

## In this notebook we will discuss *inferential statistics*

```python
import random
from scipy.stats import norm
from scipy.stats import t
import plotly.express as px
import math as mt
import numpy as np
```

These are all the libraries that we will need

```python
sample_size=31
sample_count=1000

x_values=[(sum([random.uniform(0.0,1.0) for i in range(sample_size)])/sample_size) for _ in range(sample_count)]

y_values=[1 for _ in range(sample_count)]

px.histogram(x=x_values, y=y_values, nbins=20).show()
```
