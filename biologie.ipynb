```python
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
data = [21, .04, 78, 1] 
keys = ['Sauerstoff', 'Kohlenstoffdioxid', 'Stickstoff', 'Edelgase'] 
  
palette_color = sn.color_palette('bright') 
  
plt.pie(data, labels=keys, colors=palette_color, autopct='%.0f%%') 
  
plt.title("Einatemluft")
plt.show()
```

```python
data1 = [17, 4, 78, 1] 
  
palette_color = sn.color_palette('bright') 
  
plt.pie(data1, labels=keys, colors=palette_color, autopct='%.0f%%') 
plt.title("Ausatemluft")
plt.show()
```

```python
Durchgang=np.array([1,2,3,4,5])
Anzahl_der_Atemzuege=np.array([19,23,25,22,37])
plt.scatter(Durchgang,Anzahl_der_Atemzuege)
plt.xlabel("Anzahl der Durchgänge")
plt.ylabel("Anzahl der Atemzüge")
plt.title("Atemfrequenz")
plt.show()
```
